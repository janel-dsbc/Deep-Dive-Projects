```python
#Plot US Stamp Prices over Time
import pandas as pd
import matplotlib.pyplot as plt
```

Data set source:  https://vincentarelbundock.github.io/Rdatasets/datasets.html

```python
#pd.read_csv turns a csv file into a pandas dataframe
stamp_prices = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/Stat2Data/USstamps.csv')
```

```python
#type () will return the type of an object
type(stamp_prices)
```

```python
#head() will return the column headers and first few rows of data
stamp_prices.head()
```

```python
#plot year vs price
#set a plot title first
plt.title('US Stamp Prices over Time')
#set columns to plot
plt.scatter(stamp_prices['Year'],stamp_prices['Price'])
```
